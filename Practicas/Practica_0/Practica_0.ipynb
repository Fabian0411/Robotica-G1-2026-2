## Previo de la Práctica 0

### 1. Estructura de paquetes en ROS 2
Un paquete en ROS 2 es la unidad organizativa principal del software, que permite compilar, instalar y compartir código de manera modular. Para un desarrollo basado en Python (`ament_python`), la estructura estándar que se genera dentro del espacio de trabajo (*workspace*) es la siguiente:

* **`package.xml`**: Es el archivo manifiesto. Contiene los metadatos del paquete (nombre, versión, autor, licencia) y, lo más importante, declara las dependencias necesarias para que el paquete funcione (por ejemplo, `rclpy` o `std_msgs`).
* **`setup.py`**: Es el script de instalación. Define cómo se debe instalar el paquete y declara los puntos de entrada (*entry points*), es decir, le indica a ROS 2 qué scripts de Python corresponden a los ejecutables de los nodos.
* **`resource/`**: Una carpeta generada automáticamente que contiene un archivo marcador con el nombre del paquete para el índice de recursos de ROS.
* **`p0_py/` (Directorio fuente)**: Es la carpeta principal que lleva el mismo nombre del paquete. Aquí es donde residen los scripts de Python con la lógica de nuestros nodos (en este caso, `generador_rpm.py` y `convertidor_rads.py`).

### 2. Definición de Nodos y Tópicos

#### Nodos
En la arquitectura de ROS 2, un **nodo** es un proceso ejecutable independiente que realiza una tarea de computación específica. En lugar de tener un único programa monolítico que controle todo un robot, el sistema se divide en múltiples nodos altamente especializados (ej. un nodo lee sensores, otro procesa visión, otro controla motores). Esto otorga modularidad y tolerancia a fallos. Los nodos se descubren automáticamente en la red gracias al middleware DDS (*Data Distribution Service*).

#### Tópicos
Los **tópicos** son uno de los canales de comunicación principales entre los nodos, operando bajo un modelo asíncrono de publicación/suscripción (*Publisher/Subscriber*). 
* Un nodo que genera datos (como nuestro sensor de RPM) actúa como **Publicador** y envía mensajes de un tipo específico a un tópico determinado.
* Cualquier otro nodo que necesite esa información actúa como **Subscriptor**, "escuchando" ese tópico para recibir los datos en tiempo real.
* Los tópicos desacoplan la comunicación: el publicador no necesita saber quién lo escucha, ni el subscriptor necesita saber quién envía la información.

## Desarrollo de la práctica 

El desarrollo de la práctica se divide en las siguientes partes:

### 1. Conocer las herramientas computacionales para la robótica de la clase.
El primer paso consistió en la preparación del entorno de trabajo en la Máquina Virtual con Linux. Se utilizó la terminal para inicializar las variables de entorno de ROS 2 Jazzy, lo cual es un requisito indispensable en cada nueva sesión para que el sistema reconozca los comandos base:
```bash
source /opt/ros/jazzy/setup.bash
```
### 2. Generación de un paquete de ROS 2 en Python.
Dentro del espacio de trabajo (`~/ROS2Dev/ws_sem26_2/src`), se generó la estructura base del proyecto. Se optó por Python como lenguaje de programación y se declaró la licencia MIT desde su concepción:
```bash
ros2 pkg create --build-type ament_python --license MIT p0_py
```
### 3. Desarrollo de los Nodos y Tópicos correspondientes para la práctica. 
Se programaron dos scripts principales dentro de la carpeta fuente del paquete:

* **Nodo Generador (`generador_rpm.py`):** Configurado como un publicador puro. Utiliza el reloj del sistema para generar una señal senoidal normalizada ($f(t) = 0.5 \sin(t) + 0.5$) que varía en un rango estricto de 0 a 1, simulando la velocidad de un motor en RPM. Esta señal se transmite mediante mensajes de tipo `Float64` por el tópico `/rpm` a una frecuencia de 10 Hz.
* **Nodo Convertidor (`convertidor_rads.py`):** Configurado con una arquitectura híbrida (subscriptor-publicador). Se suscribe al tópico `/rpm` recibiendo los datos en formato `Float64`, intercepta el valor y aplica la conversión matemática. Posteriormente, republica el dato procesado en el nuevo tópico `/rad_s`.
Una vez escrita la lógica, se procedió a la compilación del paquete en la raíz del *workspace*:
```bash
cd ~/ROS2Dev/ws_sem26_2
colcon build
```
### 4. El manejo de herramientas para la documentación empleando Jupyter Lab.
Finalmente, se utilizó el entorno de Jupyter (`.ipynb`) integrado en Visual Studio Code para la elaboración de este reporte interactivo. Esta herramienta permitió combinar de manera fluida el marco teórico, los bloques de código ejecutables y el formato Markdown (incluyendo ecuaciones matemáticas) para estructurar el proyecto bajo estándares académicos profesionales.

## Resultados

Se programó un Nodo Publicador que genera un mensaje de tipo `Float64` variando en un rango normalizado de $0$ a $1$. Simultáneamente, el Nodo Subscriptor-Publicador intercepta este valor y lo transforma a velocidad angular en radianes por segundo ($\frac{rad}{s}$).

La conversión matemática aplicada en el nodo subscriptor para transformar las revoluciones por minuto a velocidad angular es:

$$
w = RPM \times \frac{2\pi}{60}
$$

A continuación, se presenta la relación entre la señal de entrada (simulando RPM en un rango de 0 a 1) y la señal de salida ya convertida a $\frac{rad}{s}$, capturada en tiempo real mediante las herramientas de visualización de ROS 2.

### Resultados numéricos

Para comprobar la precisión del nodo convertidor, se evaluó el pico máximo de la señal. 

El valor máximo simulado de entrada es de $1.0$ RPM. Aplicando la transformación, el valor máximo de velocidad angular obtenido en el tópico de salida es:

$$
w = 0.1047 \frac{rad}{s}
$$

## Conclusiones

A lo largo de esta práctica, se logró el objetivo de comprender las bases de la gestión de información y la comunicación en ROS 2. Aunque al principio la arquitectura de nodos y tópicos puede parecer un poco abstracta, al programar los scripts y ver el flujo de datos en tiempo real quedó muy clara su mayor ventaja: la **modularidad**.

Como se comprobó en las gráficas y en los resultados numéricos (alcanzando el pico máximo de $w = 0.1047 \frac{rad}{s}$ de manera exacta), logramos que un nodo se dedicara exclusivamente a generar una señal simulada, mientras otro interceptaba y procesaba esa información de forma asíncrona. Esta separación de tareas es vital para el desarrollo de sistemas mecatrónicos reales, ya que permite integrar múltiples sensores y actuadores de manera escalable y tolerante a fallos.
